```python
def check_input(s):
    return isinstance(s, str)


def get_substring(text):
    current_sub = ""
    longest_sub = ""
    for char in text:
        if char in current_sub:
            idx = current_sub.find(char)
            current_sub = current_sub[idx + 1:]
        current_sub += char
        if len(current_sub) > len(longest_sub):
            longest_sub = current_sub
    return longest_sub


def main(arr):
    starting_index = -1
    longest_string = ""
    for i, item in enumerate(arr):
        if check_input(item):
            if len(item) > len(longest_string):
                longest_string = item
                starting_index = i
    result = get_substring(longest_string)
print(starting_index, result)
```
